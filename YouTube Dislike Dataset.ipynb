### 5. Convert datatype of column published_at from object to pandas datetime.

```python
# Copying the original data in new variable
YouTube_Data_copy = YouTube_Data.copy()
```

```python
# Importing datetime module
from datetime import datetime
```

```python
# Converting the published_at in datetime format
YouTube_Data_copy['published_at'] = pd.to_datetime(YouTube_Data_copy['published_at'])
```

```python
YouTube_Data_copy['published_at'].head()
```

### 6. Create a new column as 'published_month' using the column published_at (display the months only)

```python
YouTube_Data_copy['published_month'] = YouTube_Data_copy['published_at'].dt.month
```

```python
YouTube_Data_copy['published_month']
```

### 7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....

```python
replace_month = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
```

### 8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count.

```python
# Getting the count of the video with respect of published_month, storing it new variable

Total_vid_month= pd.DataFrame(YouTube_Data_copy.groupby('published_month')['video_id'].count())
```

### 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.
